# GUI FOR LOAN PREDICTION

```python
from tkinter import *
from tkinter import ttk
import joblib
import pandas as pd
```

```python
def enter_details():
    
    p1 = float(gender.get())
    p2 = float(marital.get())
    p3 = float(dependents.get())
    p4 = float(edu.get())
    p5 = float(s_employed.get())
    p6 = float(a_income.get())
    p7 = float(c_income.get())
    p8 = float(loanamount.get())
    p9 = float(lm_term.get())
    p10 = float(c_history.get())
    p11 = float(p_area.get())
    
    model = joblib.load('loan_approval_prediction')
    df = pd.DataFrame({
    'Gender':p1,
    'Married':p2,
    'Dependents':p3,
    'Education':p4,
    'Self_Employed':p5,
    'ApplicantIncome':p6,
    'CoapplicantIncome':p7,
    'LoanAmount':p8,
    'Loan_Amount_Term':p9,
    'Credit_History':p10,
    'Property_Area':p11
},index=[0])
    result = model.predict(df)
    if result == 1:
        Label(root, text="Loan approved").grid(row=31)
    else:
        Label(root, text="Loan Not Approved").grid(row=31)
        
    
root =Tk()
root.title("Loan Approval Prediction")
label = Label(root,text = "Loan Approval Prediction",bg = "white",
               fg = "red", font="bold").grid(row=0,columnspan=2)

# Gender Label
gend_label=Label(root,text = "Gender [1:Male ,0:Female]")
gend_label.grid(row=1,padx=2,pady=7)

gender = ttk.Combobox(root,state="readonly")
gender["value"] = ("0", "1")
gender.current(0)
gender.grid(row=1, column=1, padx=2, pady=7)

# Married Label
Marital_label=Label(root,text = "Married [1:Yes,0:No]")
Marital_label.grid(row=2,padx=2,pady=7)

marital = ttk.Combobox(root,state="readonly")
marital["value"] = ("0", "1")
marital.current(0)
marital.grid(row=2, column=1, padx=2, pady=7)

# Dependents Label
dep_label=Label(root,text = "Dependents")
dep_label.grid(row=3,padx=2,pady=7)

dependents = ttk.Combobox(root,state="readonly")
dependents["value"] = ("0","1","2","3")
dependents.current(0)
dependents.grid(row=3, column=1, padx=2, pady=7)

# Education Label
edu_label=Label(root,text = "Education [0:grad 1:not grad]")
edu_label.grid(row=4,padx=2,pady=7)

edu = ttk.Combobox(root,state="readonly")
edu["value"] = ("0","1")
edu.current(0)
edu.grid(row=4, column=1, padx=2, pady=7)

# Self-employed label
emp_label=Label(root,text = "Self-Employed [1:Yes 0:No]")
emp_label.grid(row=5,padx=2,pady=7)

s_employed = ttk.Combobox(root,state="readonly")
s_employed["value"] = ("0","1")
s_employed.current(0)
s_employed.grid(row=5, column=1, padx=2, pady=7)

# Applicant Income Label
appli_label=Label(root,text = "ApplicantIncome")
appli_label.grid(row=6,padx=2,pady=7)

a_income=ttk.Entry(root)
a_income.grid(row=6,column=1,padx=2,pady=7) 

# Coapplicant Income Label
coappli_label=Label(root,text = "CoapplicantIncome")
coappli_label.grid(row=7,padx=2,pady=7)

c_income=ttk.Entry(root)
c_income.grid(row=7,column=1,padx=2,pady=7) 

# Loan amount Label
l_amou_label=Label(root,text = "LoanAmount")
l_amou_label.grid(row=8,padx=2,pady=7)

loanamount=ttk.Entry(root)
loanamount.grid(row=8,column=1,padx=2,pady=7)
        
# Loan amount term Label
l_amou_term_label=Label(root,text = "Loan_Amount_Term")
l_amou_term_label.grid(row=9,padx=2,pady=7)

lm_term=ttk.Entry(root)
lm_term.grid(row=9,column=1,padx=2,pady=7)


# Credit History Label
c_hist_label=Label(root,text = "Credit_History")
c_hist_label.grid(row=10,padx=2,pady=7)

c_history = ttk.Combobox(root,state="readonly")
c_history["value"] = ("0","1")
c_history.current(0)
c_history.grid(row=10, column=1, padx=2, pady=7)


# Property Area Label
prop_label=Label(root,text = "Property_Area [0: Rural,1: Urban,2:Semiurban]")
prop_label.grid(row=11,padx=2,pady=7)

p_area = ttk.Combobox(root,state="readonly")
p_area["value"] = ("0","1","2")
p_area.current(0)
p_area.grid(row=11, column=1, padx=2, pady=7)

# Prediction button
Button(root,text="Predict",command=enter_details).grid()

mainloop()
```
